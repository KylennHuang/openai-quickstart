### 1. 英文转换为中文

> python ai_translator/main.py --model_type OpenAIModel --openai_api_key $OPENAI_API_KEY --file_format markdown --book tests/test.pdf --openai_model gpt-3.5-turbo




### 2. 英文转换为法语

> python ai_translator/main.py --target_language '韩语' --model_type OpenAIModel --openai_api_key $OPENAI_API_KEY --file_format markdown --book tests/test.pdf --openai_model gpt-3.5-turbo




### 3. 中文转换为日语

> python ai_translator/main.py --target_language '日语' --model_type OpenAIModel --openai_api_key $OPENAI_API_KEY --file_format markdown --book tests/kongyiji.pdf --openai_model gpt-3.5-turbo
